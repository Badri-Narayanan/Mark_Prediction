```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import pickle
```

```python
X = pd.read_csv("../../Data Sets/CB Challenge/Challenge1/Linear_X_Train.csv")
y = pd.read_csv("../../Data Sets/CB Challenge/Challenge1/Linear_Y_Train.csv")
print(X.shape)
print(y.shape)
```

```python
learning_rate = LinearRegression(normalize = True)
```

```python
learning_rate.fit(X, y)
```

```python
plt.scatter(X, y)
plt.title("Actual Observation (Training Points)")
plt.xlabel("Time Spent to Study")
plt.ylabel("Marks Obtained")
plt.grid(True)
plt.show()
```

```python
y_pred = learning_rate.predict(X)
y_pred
```

```python
plt.scatter(X, y)
plt.plot(X, y_pred, color = "orange", label="Prediction Line")
plt.title("Actual Observation (Training Points) and Predicted Line (Best Fit Line)")
plt.xlabel("Time Spent to Study")
plt.ylabel("Marks Obtained")
plt.legend()
plt.grid(True)
plt.show()
```

```python
learning_rate.score(X, y)
```

```python
learning_rate.predict([[10]])
```

```python
pickle.dump(learning_rate, file = open("model.pkl", 'wb+'))
```
